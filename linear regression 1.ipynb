```python
sex = pd.get_dummies(df['sex'],drop_first = True)
df = pd.concat([df,sex],axis = 1)
df.info()
```

```python
smoke = pd.get_dummies(df['smoker'],drop_first = True)
df = pd.concat([df,smoke],axis = 1)
df.info()
```

```python
region = pd.get_dummies(df['region'],drop_first = False)
df = pd.concat([df,region],axis = 1)
df.info()
```

```python
df.head()
```

```python
df = df.drop(['sex','smoker','region'], axis= 1)
df.head()
```

```python
x = df.drop(['charges'], axis =1)
y = df['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 100)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
im = LinearRegression()
im.fit(x_train, y_train)
```

```python
print(im.intercept_)
```

```python
coeffecients = pd.DataFrame(im.coef_, x.columns)
coeffecients.columns=['Coeffecients']
coeffecients
```

```python
predictions = im.predict(x_test)
plt.scatter(y_test, predictions)
```

```python
sns.histplot((y_test-predictions), kde = True, bins=40)
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
print('R2 Score for LinearRegression on test sata:{}'.format(np.round(r2_score(y_test, predictions), 3)))
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
